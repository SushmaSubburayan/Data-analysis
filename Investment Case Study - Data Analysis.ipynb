```python
import numpy as np
import pandas as pd

#Reading companies.txt file by using encoding = "ISO-8859-1 
companies = pd.read_csv("D:/Sushma/Data Science/Investment Case Study/companies.txt", sep="\t", encoding = "ISO-8859-1")
companies.head()
```

```python
#Reading rounds2.csv file by using encoding = "ISO-8859-1" and storing the data in rounds2 data frame
rounds2 = pd.read_csv("D:/Sushma/Data Science/Investment Case Study/rounds2.csv", sep=",", encoding = "ISO-8859-1")
rounds2.head()
```

```python
#Changing permalink field to lower case
companies['permalink']=companies.permalink.str.lower()

# Decoding to read special characters
companies['permalink']=companies.permalink.str.encode('utf-8').str.decode('ascii','ignore')

#Extracting the unique comapnies from companies data frame
companies_unique_companies=companies['permalink'].unique().tolist()

#Displaying the top 5 values 
companies.head()
```

```python
#Changing company permalink field to lower case
rounds2['company_permalink']=rounds2.company_permalink.str.lower()

# Decoding to read any special character
rounds2['company_permalink']=rounds2.company_permalink.str.encode('utf-8').str.decode('ascii','ignore')

#Extracting the unique companies from rounds2 data frame
rounds2_unique_companies=rounds2['company_permalink'].unique().tolist()

#Displaying top 5 values
rounds2.head()
```

```python
#Number of unique companies in rounds2
len(rounds2_unique_companies)
```

```python
#Number of unique companies in companies file
len(companies_unique_companies)
```

```python
#Companies in the rounds2 file which are not  present in companies
set(rounds2_unique_companies).difference(set(companies_unique_companies))
```

```python
#Merging the companies frame and the rounds2 data frame and creating the master_frame.
master_frame = pd.merge(rounds2,companies,how='left',left_on='company_permalink', right_on='permalink')

#Number of observations present in master_frame
len(master_frame)
```

```python
#Inspecting the master dataframe created 
master_frame.info()
master_frame.shape
```

```python
#Displaying top 5 values
master_frame.head()
```

```python
#Calculating the percentage of null values present in all the columns
print(round(100*(master_frame.isnull().sum()/len(master_frame.index)),2))
```

```python
#Dropping the columns with high percentage of null values
master_frame.drop(['funding_round_code','founded_at','homepage_url','state_code','region','city'],axis=1,inplace=True)
```

```python
#Calculating the percentage of null values present in each column after dropping the above columns
print(round(100*(master_frame.isnull().sum()/len(master_frame.index)),2))
```

```python
#Dropping the rows with null values for raised_amount_usd and country_code
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]
master_frame = master_frame[~pd.isnull(master_frame['country_code'])]
```

```python
#Calculating the percentage of null values present in each column after dropping rows with high null values
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#Retaining only those rows of 'category_list' column which doesn't have null values
master_frame = master_frame[~pd.isnull(master_frame['category_list'])]
```

```python
#Calculating the percentage of null values present in each column after dropping rows with high null values
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#Converting the unit of raised_amount_usd to million USD for ease of understanding and calculation 
master_frame.raised_amount_usd=master_frame.raised_amount_usd/1000000

#Calculating the average investment amount for each of the funding types in million USD
avg_funding = master_frame.groupby(['funding_round_type']).mean()

avg_funding['raised_amount_usd']
```

```python
#Finding the average funding for venture, angel, seed, and private equity funding types
master_frame.loc[master_frame['funding_round_type'].isin(['venture', 'angel', 'seed','private_equity'])].groupby(['funding_round_type']).mean()
```

```python
#Finding the companies which received the average funding between 5 to 15 million USD 
avg_funding.loc[(avg_funding['raised_amount_usd']>=5.0) & (avg_funding['raised_amount_usd']<=15.0)]
```

```python
#Selecting only those rows which have 'venture' as the investment type
total_funding_received = master_frame.loc[master_frame['funding_round_type']=='venture']

#Getting top nine countries which have received the highest total funding for 'venture' investment type
top9 = (total_funding_received.groupby(['country_code']).sum())['raised_amount_usd'].sort_values(ascending=False).head(9)
top9
```

```python
#Loading mapping.csv file in categories dataframe and displaying top 5 values
categories = pd.read_csv("D:/Sushma/Data Science/Investment Case Study/mapping.csv")
categories.head()
```

```python
#Dropping the rows with category list as null or NaN and displaying values
categories = categories[~pd.isnull(categories['category_list'])]
categories.head()
```

```python
#Extracting the primary sector of each category list from the category_list column and adding as new column in master frame 
master_frame['category_list'].astype(str)
category=master_frame['category_list'].str.split("|",expand=True)
master_frame['primary_sector']=category[0]
```

```python
#Checking if primary sector column is added 
master_frame.columns
```

```python
#Displaying the top 5 rows in the master frame to check the values in primary_sector column
master_frame.head()
```

```python
#Creating a new dataframe by mapping each categoring to its main sector

category_mapping=categories.melt(id_vars=['category_list'], var_name='main_Sector')
category_mapping=category_mapping.loc[category_mapping.value==1]
category_mapping.drop('value',axis=1,inplace=True)
category_mapping.head()
```

```python
#Merging master frame to category_mapping to create merged dataframe on categorylist

master_frame=pd.merge(master_frame,category_mapping,how='left',left_on='primary_sector',right_on='category_list')
master_frame.head()
print(round(100*(master_frame.isnull().sum()/len(master_frame.index)),2))
```

```python
#Dropping the rows with null values in the category_list_y column
master_frame = master_frame[~pd.isnull(master_frame['category_list_y'])]
```

```python
#Dropping the rows with null values in the main_sector column
master_frame = master_frame[~pd.isnull(master_frame['main_Sector'])]
```

```python
# Creating D1 , D2 and D3 dataframes based on the investment constraints specified by Spark Funds for top 3 English speaking countries 
# USA,GBR and IND with funding round type as venture and investement from 5 to 15 million $

D1=master_frame.loc[(master_frame.country_code=='USA')& (master_frame.funding_round_type=='venture')& (master_frame.raised_amount_usd<=15.0)& (master_frame.raised_amount_usd>=5)]
D2=master_frame.loc[(master_frame.country_code=='GBR')& (master_frame.funding_round_type=='venture')& (master_frame.raised_amount_usd<=15.0)& (master_frame.raised_amount_usd>=5)]
D3=master_frame.loc[(master_frame.country_code=='IND')& (master_frame.funding_round_type=='venture')& (master_frame.raised_amount_usd<=15.0)& (master_frame.raised_amount_usd>=5)]
    
```

```python
#Adding two more columns to D1 for sum of amount raised per primary sector and count of companies per main sector
D1['raised_per_sector'] = D1.groupby(['main_Sector'])['raised_amount_usd'].transform(sum)
D1['count_per_sector'] = D1.groupby(['main_Sector'])['primary_sector'].transform('count')
D1.head()
```

```python
#Adding two more columns to D2 for sum of amount raised per primary sector and count of companies per main sector
D2['raised_per_sector'] = D2.groupby(['main_Sector'])['raised_amount_usd'].transform(sum)
D2['count_per_sector'] = D2.groupby(['main_Sector'])['primary_sector'].transform('count')
D2.head()
```

```python
#Adding two more columns to D3 for sum of amount raised per primary sector and count of companies per main sector
D3['raised_per_sector'] = D3.groupby(['main_Sector'])['raised_amount_usd'].transform(sum)
D3['count_per_sector'] = D3.groupby(['main_Sector'])['primary_sector'].transform('count')
```
